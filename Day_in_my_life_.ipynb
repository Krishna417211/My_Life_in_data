```python
#Importing the files
#New project My daily routine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
upload = files.upload()
```

```python
#Now reading the csv file
df = pd.read_csv('/content/Day in life.txt')
#Understanding the datatypes and shapes of the given Dataset
print(df.head())
print(df.dtypes)
print(df.shape)
```

```python
a = df.iloc[:,1:]
a.to_numpy()
total_hours = np.sum(a)
total_average = np.mean(a)
```

```python
#Making a pie chart of total hours spend per activity

labels = total_hours.index
values = total_hours.values

plt.pie(values,labels = labels,autopct='%1.1f%%', startangle=90)
```

```python
#Making a line graph of no of hours study daily
plt.plot(df['Date'],df['Study'])
plt.xlabel('Date')
plt.ylabel('Study')
plt.title('No of hours I study')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Making a new column Productive hours
df['Productive hours'] = df['Sleep'] + df['Study'] + df['Exercise'] + df['Chores']
b = df['Productive hours'].to_numpy()
# The Maximum productive hours
int(b.max())
# The Minimum productive hours
int(b.min())
# The Average of the productive hours
int(b.mean())

# To find the date on which max productivity occured
MP = df.loc[df['Productive hours'].idxmax(),'Date']
# To find the data on which min productivity occured
LP = df.loc[df['Productive hours'].idxmin(),'Date']
```

```python

# Step 1: Extract required data
activities = df[['Date', 'Sleep', 'Study', 'Exercise']]
x = np.arange(len(df))  # positions for each date
width = 0.25  # width of each bar

# Step 2: Plot grouped bars
plt.figure(figsize=(12, 6))
plt.bar(x - width, activities['Sleep'], width, label='Sleep')
plt.bar(x,         activities['Study'], width, label='Study')
plt.bar(x + width, activities['Exercise'], width, label='Exercise')

# Step 3: X-axis & Labels
plt.xticks(x, activities['Date'], rotation=45)
plt.ylabel("Hours")
plt.title("Daily Sleep, Study & Exercise Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# +1 → Strong Positive (both go up together)

# -1 → Strong Negative (one goes up, other goes down)

# 0 → No relation

# the plot showing the relation between the daily works

Corr = df[['Sleep','Study','Exercise','Chores']].corr()
sns.heatmap(Corr,annot = True,cmap = 'RdBu',vmin =-1,vmax = 1)
plt.title('The Correlation between daily activities')
plt.show()

```
